## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
df = df.dropna(subset=['name', 'host_name']).copy()
df.loc[:, 'reviews_per_month'] = df['reviews_per_month'].fillna(0)

df.loc[:, 'last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
df = df[df['price'] > 0]
df = df[df['price'] <= 1000]
df = df[df['minimum_nights'] <= 365]

df.head()
````

### What all manipulations have you done and insights you found?

Answer: Handled Missing Values - Removed the row where name or host_name was missing. Filled missing reviews_per_month with 0. Converted last_review to proper datetime format.

Removed Invalid and Outlier Values - Removed the rows where price was 0. Removed the rows where price is more than 1000. Remove the rows with minimum nights was greater than 365.

The insights that we get was most of the price is under 200 but there are few outlier with price more than 1000. Most of the minimum nights are 1 to 3 days and there are few outliers. Many of the rows have 0 reviews.

#### Chart - 1: Listings Count by Neighbourhood Group(Count Plot)

````python
# Chart - 1 visualization code
sns.countplot(data=df, x='neighbourhood_group', order=df['neighbourhood_group'].value_counts().index)
plt.title("Number of Listings by Neighbourhood Group")
plt.xlabel("Neighbourhood Group")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

Answer: We want to compare the frequency of listings in each case. A bar chart for categorical data is ideal to show how listings are distributed.

##### 2. What is/are the insight(s) found from the chart?

#### Chart - 11 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
numeric_cols = ['price', 'minimum_nights', 'number_of_reviews',
                'reviews_per_month', 'calculated_host_listings_count', 'availability_365']

corr = df[numeric_cols].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numeric Features")
plt.tight_layout()
plt.show()
````

#### Chart - 12 - Pair Plot
